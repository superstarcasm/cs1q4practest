```python
#This code finds the average of the 3 test scores given by the student user.
#It then decides if the student has passed or failed.
#Additionally, it checks for if an impossible to achieve score has been inputted.
def AVERAGE():
  return (SCORE1 + SCORE2 + SCORE3)/3

print("Hello!")
print(" ")
NAME = input("Please input your name.")
print(" ")
print("Welcome,", NAME, ",to the Student Score Analyzer!")
print(" ")

SCORE1 = float(input("Please input your first score."))
print(" ")

SCORE2 = float(input("Please input your second score."))
print(" ")

SCORE3 = float(input("Please input your third score."))
print(" ")

print(" ")
print("Thank you. Now calculating your scores..")
print(" ")
if AVERAGE() > 15 and AVERAGE() < 25:
  print("You have passed with an average score of", AVERAGE(), "!")
if AVERAGE() < 15:
  print("You have failed with an average score of", AVERAGE(), ".")
if SCORE1 > 25 or SCORE2 > 25 or SCORE3 > 25:
  if SCORE1 < 0 or SCORE2 < 0 or SCORE3 < 0:
    print("You have inputted invalid or impossible scores. Please, try again.")
print(" ")
print("Thank you for using the Student Score Analyzer!")

```
